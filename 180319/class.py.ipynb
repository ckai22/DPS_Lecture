```python
g = loadtxt("pop.txt")
```

```python
plt.xlabel('Day')
plt.ylabel('Population density')
plt.plot(g[:,0],g[:,1],'ms')
plt.plot(g[:,0],g[:,1],'m')
```

```python
k = g[:,0]
n = g[:,1]
```

```python
dn = n[1:]-n[0:-1]
```

```python
plt.plot(k[:-1],dn)
```

```python
plt.plot(n[:-1],dn,'.')
```

```python
m = linspace(0, 650, 100)
print(m)
```

```python
dm = 0.0015*(550 - m)*m
```

```python
plt.plot(n[:-1],dn,'.')
plt.plot(m, dm)
```

```python
nm = zeros(len(k))
nm[0] = n[0]
for i in range(1, len(k)):
    nm[i] = nm[i-1]*(1 + 0.0015*550 - 0.0015*nm[i-1])
```

```python
plt.plot(g[:,0],g[:,1],'s')
plt.plot(k,nm)
```

```python
x0 = 0.3
kmax = 500
kc = 300
M = 1000
r =linspace(0, 4, M)
x = r*(1-x0)*x0
for i in range(kmax):
    x = r*(1-x)*x
    if i >= kc:
        plt.plot(r, x, 'm.')
```

```python
r = 3.64
x1 = 0.3
x2 = 0.3 + 10**(-10)
o1= []
o2 = []
for i in range(250):
    x1 = r*(1-x1)*x1
    x2 = r*(1-x2)*x2
    if i >= 200:
        o1 += [x1]
        o2 += [x2]
plt.plot(o1)
plt.plot(o2)

#chaotic하다는 것과 random하다는 것은 완전히 다른 이야기.
#앞에 있는 history로 후의 일을 전혀 예측할 수 없을 때. Random이라고 함
```
