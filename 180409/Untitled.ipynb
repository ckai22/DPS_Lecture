```python
import numpy as np
import random as ran
from matplotlib.pyplot import *
from mpl_toolkits.mplot3d import Axes3D
import math
```

```python
particleX = np.zeros(100)
particleY = range(0,50)
stepAvg = np.zeros(50)
stepVar = np.zeros(50)
```

```python
for k in range(50):
    for i in range(0, 100):
        num = ran.random()
        if(num > 0.5):
            particleX[i] += 1
        else:
            particleX[i] += -1
    stepAvg[k] = particleX.mean()
    stepVar[k] = particleX.var()
    
```

```python
print(particleX)
print(stepAvg)
print(stepVar)
```

```python
plt.xlim(-100, 100)
plt.plot(particleX, particleY,".")
```

```python
plot(particleY, stepAvg,"b")
plot(particleY, stepVar,"r")
```

```python
px = np.zeros(100)
py = np.zeros(100)
px2 = np.zeros(100)
py2 = np.zeros(100)
```

```python
for k in range(50):
    for i in range(100):
        r = ran.random()
        r2 = ran.random()
        if(r < 0.25):
            py[i] = py[i] + 1
            py2[i] += 4
        elif(r < 0.5):
            px[i] = px[i] + 1
            px2[i] += 4
        elif(r < 0.75):
            py[i] = py[i] - 1
            py2[i] -= 4
        else:
            px[i] -= 1
            px2[i] -= 4
```

```python

plt.plot(px, py, "ro")
plt.plot(px2, py2, "bo")
```
